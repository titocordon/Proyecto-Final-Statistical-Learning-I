### Visualizaci√≥n

```python
# Diagrama de sectores para una columna categ√≥rica
plt.figure(figsize=(8, 8))
df['Make'].value_counts().plot(kind='pie', autopct='%1.1f%%', wedgeprops=dict(width=0.3))
plt.title('Diagrama de sectores para variable Make')
plt.ylabel('')
plt.show()
```

```python
# Gr√°fico de barras 

fig = plt.figure(figsize=(10,6))
sns.countplot( x='Model Year', data=df)
plt.title('Gr√°fico de barras')
plt.show()
```

```python
# Diagrama de sectores para una columna categ√≥rica
plt.figure(figsize=(8, 8))
df['Electric Vehicle Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', wedgeprops=dict(width=0.3))
plt.title('Diagrama de sectores para variable Electric Vehicle Type')
plt.ylabel('')
plt.show()
```

```python
# Diagrama de sectores para una columna categ√≥rica
plt.figure(figsize=(8, 8))
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts().plot(kind='pie', autopct='%1.1f%%', wedgeprops=dict(width=0.3))
plt.title('Diagrama de sectores para variable Clean Alternative Fuel Vehicle (CAFV) Eligibility')
plt.ylabel('')
plt.show()
```
